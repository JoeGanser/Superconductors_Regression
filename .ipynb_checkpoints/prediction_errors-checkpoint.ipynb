```python
import random
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()



df = pd.read_csv('data_filtered.csv').drop('Unnamed: 0',axis=1)
#df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
X = df.drop('critical_temp',axis=1)
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)

corr = pd.DataFrame(df.corr()['critical_temp'])
corr['abs'] = np.abs(corr['critical_temp'])
corr = corr.sort_values(by='abs',ascending=False).drop('abs',axis=1).dropna()
corr = corr.reset_index().rename(columns={'index':'feature','critical_temp':'correlation'})
corr = corr.loc[1:]



y = df['critical_temp']



rf = RandomForestRegressor(n_estimators=150,max_features='sqrt')
e=0.1515151515151516
intercept = 1.5542
y_t = y.apply(lambda x: (x**e)-intercept)
X_train, X_test, y_train, y_test = train_test_split(X, y_t, test_size=0.3, random_state=random.randint(1,101))

rf.fit(X_train,y_train)

y_test = y_test.apply(lambda x: (x+intercept)**(1/e))
y_train = y_train.apply(lambda x: (x+intercept)**(1/e))


y_pred_test = pd.Series(rf.predict(X_test)).apply(lambda x: (x+intercept)**(1/e))
y_pred_train = pd.Series(rf.predict(X_train)).apply(lambda x: (x+intercept)**(1/e))
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_test))
```

```python
from scipy import stats



slope, intercept, r_value, p_value, std_err = stats.linregress(y_test,y_pred_test)
line = slope*y_test+intercept
plt.figure(figsize=(5,5))
plt.plot(y_test,line,color='red')
plt.scatter(y_test,y_pred_test)
plt.ylabel('Predicted Temp')
plt.xlabel('Actual Temp')
plt.title('Actual Price vs Predicted')
print("If you did it correctly, your plot should looks like this:")
plt.show()
```

```python
import seaborn as sns
from scipy.stats import norm
from scipy import stats
y_pred_test = y_pred_test.dropna()
y_pred_train = y_pred_train.dropna()
y_test = y_test.dropna()
y_train = y_train.dropna()

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.title('Errors on test set - Distplot')
sns.distplot((y_pred_test-y_test).dropna(), fit=norm)
plt.xlabel('')

plt.subplot(2,2,2)
plt.title('Errors on train set - Distplot')
sns.distplot((y_pred_train-y_train).dropna(), fit=norm)
plt.xlabel('')


plt.subplot(2,2,3)
res = stats.probplot((y_pred_test-y_test).dropna(), plot=plt)
plt.title('Q-Q plot of errors on test set')
plt.xlabel('')
plt.ylabel('')

plt.subplot(2,2,4)
res = stats.probplot((y_pred_train-y_train).dropna(), plot=plt)
plt.title('Q-Q plot of errors on train set')
plt.xlabel('')
plt.ylabel('')
plt.subplots_adjust(hspace=0.5)
plt.savefig('images/Errors_distplot.png')
plt.show()
```
